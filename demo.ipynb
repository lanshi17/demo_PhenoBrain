## 单个模型调用demo

```python
def model_name(m):
    return getattr(m, "name", getattr(m, "model_name", m.__class__.__name__))

hpo_list = ["HP:0001913", "HP:0008513", "HP:0001123"]
topk = 5
target_model_name = "MICAModel"   # 也可换成 ICTODQAcross-Ave-Random / HPOProbMNB-Random 等


models = build_available_models()
name_to_model = {model_name(m): m for m in models}

if target_model_name not in name_to_model:
    raise RuntimeError(f"{target_model_name} 不可用，当前可用: {list(name_to_model.keys())}")

```

```python
single_model = name_to_model[target_model_name]
results = single_model.query(hpo_list, topk)
print(results)
```
